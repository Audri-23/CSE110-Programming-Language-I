```python
#task 16

inp=input().split(", ")
lst=[]
max1=0
max2=0
count=0
for i in inp:
  lst.append(int(i))
print("My list:",lst)
for j in lst:
  if(j>max1):
    max1=j
for k in lst:
  if(k==max1):
    continue
  elif(k>max2):
    max2=k
for m in lst:
  if(m==max2):
    break
  count+=1
print("Second largest number in the list is",max2,"which was found at index",count)
```

```python
#task 17

inp=input().split(", ")
lst=[]
max=0
min=0
count1=0
count2=0
for i in inp:
  lst.append(int(i))
for j in lst:
  if(j<min):
    min=j
for k in lst:
  if(k==min):
    break
  count1+=1
print("Smallest number in the list is",min,"which was found at index",count1)
for l in lst:
  if(l>max):
    max=l
for m in lst:
  if(m==max):
    break
  count2+=1

print("Largest number in the list is",max,"which was found at index",count2)
```
