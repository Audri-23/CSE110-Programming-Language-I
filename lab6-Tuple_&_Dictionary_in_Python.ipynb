```python
#task 7

dict1={'Harry':15, 'Draco':8, 'Nevil':19}
dict2={'Ginie':18, 'Luna': 14}
dict3={}
for i,j in dict1.items():
  dict3[i]=j
for k,l in dict2.items():
  dict3[k]=l
print(dict3)
```

```python
#task 8

dict1={}
num=int(input("Enter the amount of key: "))
sum=0
for i in range(num):
  name=input()
  marks=int(input())
  dict1[name]=marks
for j in dict1.values():
  sum+=j
avg=sum/num
print("Average is:",int(avg))
```

```python
#task 11

string=input().lower()
dict1={}
for i in string:
  if(i==" "):
    continue
  if(i not in dict1):
    dict1[i]=1
  else:
    dict1[i]=dict1[i]+1
print(dict1)
```

```python
#task 13

list_1 = [("a", 1), ("b", 2), ("a", 3), ("b", 1), ("a", 2), ("c", 1)]
dict1={}

for i in list_1:
  A=i[0]
  B=i[1]
  if(A not in dict1):
    dict1[A]=[B]
  else:
    dict1[A].append(B)
print(dict1)
```

```python
#task 14

dict1 = {'a':59 , 'b':-82 , 'c':5 , 'd':-81 , 'e':53}
for i in dict1:
    j = 0
    k = 22
    while j < 5:
        if j % 2 == 0:
            k = dict1[i] + j - (8 + k % 6) / 3
            dict1[i] = dict1[i]+ int(k)
        else:
            k = dict1[i] + j - (6 - k % 8) * 3
            dict1[i] = dict1[i] - int(k)
        j += 1
    print(int(k))
    print(i + " -> " + str(dict1[i]))
```
