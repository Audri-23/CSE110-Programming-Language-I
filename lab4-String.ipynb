```python
#task 4

word=input("Enter a word: ")
length=len(word)

if(length<4):
   print(word)
elif(length>3):
  if(word.endswith("er")):
    print(word[:-2]+"est")
  elif(word.endswith("est")):
    print(word)
  else:
    print(word+"er")
```

```python
#task 4 alternative(without endswith)

word=input("Enter a word: ")
length=len(word)

if(length<4):
   print(word)
elif(length>3):
  if(word[length-2:]=="er"):
    print(word[:-2]+"est")
  elif(word[length-3:]=="est"):
    print(word)
  else:
    print(word+"er")
```

```python
#task 10

word=input()
num_1=0
num_2=0
length=len(word)
temp=""

while (word[num_1]!=","):
  num_1+=1
num_3=num_1+2

while(num_2<num_1 and num_3<length):
  temp+=word[num_2]+word[num_3]
  num_2+=1
  num_3+=1

if(num_2<num_1):
  temp+=word[num_2:num_1]
else:
  temp+=word[num_3:length]
print(temp)
```

```python
#task 13

test = ""
i = 5
j = 0
k = 15
while i< 10:
    k-=1
    j = k
    while j > 10:
        if (j % 2) == 0:
            test = "<--"
            test = str(test) + str(i) + str(2) + "-->" + str(int(j / 2))
        else:
            test = "-->"
            test = "-->" + str(int(i / 2)) + str(test) + str(j)
        print(test)
        j = j-1
    i+=1
```
