```python
number=int(input("Enter a number: "))
count=0
while(number!=0):
  digit=number%10
  for i in range(1,digit+1):
    if(digit%i==0):
      count+=1
  if(count==2):
    print(digit,"is a prime number")
  else:
    print(digit,"is not a prime number")
  number=number//10
  count=0
```
