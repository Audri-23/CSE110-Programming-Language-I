```python
#task 7

def show_palindrome(x):
  k=0
  for i in range(1,x+1):
    k=print(i,end="")
  for j in range(x-1,0,-1):
    k=print(j,end="")
  return k
show_palindrome(5)
```

```python
#task 8

def show_palindromic_triangle(num):
  for i in range(1,num+1):
    for j in range(num-i,0,-1):
      print(" ",end=" ")
    for k in range(1,i+1):
      print(k,end=" ")
    for l in range(i-1,0,-1):
      print(l,end=" ")
    print()
show_palindromic_triangle(5)
```

```python
#task 11

def rem_duplicate(x):
  lst=[]
  for i in x:
    if(i not in lst):
      lst.append(i)
  return tuple(lst)
print(rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2])))
```

```python
#task 16

inp=int(input())
def splitting_money(x):
  temp=""
  y=x//500
  if(y!=0):
    temp+="500 taka: "
    temp+=str(y)
    temp+=" note(s)\n"
    x=x%500
  z=x//100
  if(z!=0):
    temp+="100 taka: "
    temp+=str(z)
    temp+=" note(s)\n"
    x=x%100
  a=x//50
  if(a!=0):
    temp+="50 taka: "
    temp+=str(a)
    temp+=" note(s)\n"
    x=x%50
  b=x//20
  if(b!=0):
    temp+="20 taka: "
    temp+=str(b)
    temp+=" note(s)\n"
    x=x%20
  c=x//10
  if(c!=0):
    temp+="10 taka: "
    temp+=str(c)
    temp+=" note(s)\n"
    x=x%10
  d=x//5
  if(d!=0):
    temp+="5 taka: "
    temp+=str(d)
    temp+=" note(s)\n"
    x=x%5
  e=x//2
  if(e!=0):
    temp+="2 taka: "
    temp+=str(e)
    temp+=" note(s)\n"
    x=x%2
  f=x//1
  if(f!=0):
    temp+="1 taka: "
    temp+=str(f)
    temp+=" note(s)"
  return temp
print(splitting_money(inp))
```

```python
#task 18


inp1=int(input("Enter starting value: "))
inp2=int(input("Enter ending value: "))
inp3=int(input("Enter first divisor: "))
inp4=int(input("Enter second divisor: "))

def function_name(starting_value,ending_value,first_divisor,second_divisor):
  sum=0
  for i in range(starting_value,ending_value):
    if(i%first_divisor==0 and i%second_divisor==0):
      continue
    elif(i%first_divisor==0 or i%second_divisor==0):
      sum+=i
  return sum
print(function_name(inp1,inp2,inp3,inp4))
```
